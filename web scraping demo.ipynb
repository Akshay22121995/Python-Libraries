```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.imdb.com/search/title/?release_date=2021&sort=num_votes,desc&page=1"
response = requests.get(url)
print(response.text[:500])
```

```python
soup = BeautifulSoup(response.text,'html.parser')
type(soup)
```

```python
links = soup.find_all('div', class_ = 'lister-item mode-advanced')
links
#print(type(links))
```

```python
names=[]
mymoviedata={}
years=[]
for data in links:
    name=data.h3.a.text
    names.append(name)
    yr=data.h3.find('span', class_ = 'lister-item-year text-muted unbold')
    years.append(yr.text)
```

```python
import pandas as pd    
mymoviedata['names'] = names
mymoviedata['years'] = years
print(mymoviedata)
df=pd.DataFrame(mymoviedata)
```

```python
#to find data for all movies
# Lists to store the scraped data in
names = []
years = []
imdb_ratings = []
metascores = []
votes = []
```

```python
# Extract data from individual movie container
for container in links:
# If the movie has Metascore, then extract:
    if container.find('div', class_ = 'ratings-metascore') is not None:
# The name
        name = container.h3.a.text
        names.append(name)
    # The year
        year = container.h3.find('span', class_ = 'lister-item-year').text
        years.append(year)
    # The IMDB rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
    # The Metascore
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
    # The number of votes
        vote = container.find('span', attrs = {'name':'nv'})['data-value']
        votes.append(int(vote))
```
